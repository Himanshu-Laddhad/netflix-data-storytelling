This notebook analyzes Netflix data focusing on **Rajiv Chilaka** and **Martin Scorsese**.

```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "netflix_data.csv"
df = pd.read_csv(file_path)

# Basic overview
df.info()
df.head()
```

## Step 1: Data Cleaning

```

# Check missing values and duplicates
print("Missing values per column:")
print(df.isnull().sum())

print("\nDuplicate rows:", df.duplicated().sum())

# Handle missing values where appropriate
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")

# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

## Step 2: Filter for Rajiv Chilaka and Martin Scorsese

```

chilaka_df = df[df['director'].str.contains("Rajiv Chilaka", case=False, na=False)]
scorsese_df = df[df['director'].str.contains("Martin Scorsese", case=False, na=False)]

print("Rajiv Chilaka count:", len(chilaka_df))
print("Martin Scorsese count:", len(scorsese_df))

chilaka_df[['title','type','release_year','listed_in','duration']].head()
```

## Step 3: Comparative Analysis

```

# Movies vs TV Shows
chilaka_type = chilaka_df['type'].value_counts()
scorsese_type = scorsese_df['type'].value_counts()

print("Chilaka type distribution:\n", chilaka_type)
print("\nScorsese type distribution:\n", scorsese_type)

# Genre distribution
print("\nChilaka genres:\n", chilaka_df['listed_in'].value_counts().head())
print("\nScorsese genres:\n", scorsese_df['listed_in'].value_counts().head())

# Release year trends
plt.figure(figsize=(10,5))
sns.histplot(chilaka_df['release_year'], bins=15, label="Rajiv Chilaka", color="orange", kde=False)
sns.histplot(scorsese_df['release_year'], bins=15, label="Martin Scorsese", color="blue", kde=False)
plt.legend()
plt.title("Release Year Distribution")
plt.show()
```

## Step 4: One Killer Figure

```

# Example: Movies vs TV Shows comparison
fig, ax = plt.subplots(1,2, figsize=(12,5))

chilaka_df['type'].value_counts().plot(kind='bar', ax=ax[0], color='orange')
ax[0].set_title("Rajiv Chilaka – Movies vs TV Shows")
ax[0].set_ylabel("Count")

scorsese_df['type'].value_counts().plot(kind='bar', ax=ax[1], color='blue')
ax[1].set_title("Martin Scorsese – Movies vs TV Shows")
ax[1].set_ylabel("Count")

plt.suptitle("Key Comparison of Directors")
plt.show()
```

- **Martin Scorsese**: Focus on auteur-driven, critically acclaimed films. Strategy → Position human storytelling against GenAI.
